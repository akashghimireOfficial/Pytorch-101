```python
import torch
import torch.nn as nn
import numpy as np
from sklearn import datasets
from torch.optim import SGD
import matplotlib.pyplot as plt

```

```python
# 0) Prepare data
X_numpy, y_numpy = datasets.make_regression(n_samples=100, n_features=1, noise=20, random_state=4)
```

```python
## Preprocessing the datasets 

X=torch.from_numpy(X_numpy).to(torch.float32)
y=torch.from_numpy(y_numpy).to(torch.float32)
y=y.reshape(-1,1)
```

```python
n_samples,n_features=X_numpy.shape
n_labels=y.shape[1]
```

```python
model=nn.Linear(n_features,n_labels)
```

```python
loss_fnc=nn.MSELoss()
num_epochs=1000
learning_rate=0.01
opt=SGD(model.parameters(),lr=learning_rate)
```

```python
for epoch in range(num_epochs):
    y_pred=model(X)
    loss=loss_fnc(y,y_pred)
    loss.backward()
    opt.step() ## Updated the model parameters 
    opt.zero_grad()

    if epoch%10==0:
        print('Epoch Num: {}'.format(epoch))
        print('Loss: {:.3f}'.format(loss.item()))
```

```python
predicted=model(X).detach().numpy()
```

```python

plt.plot(X_numpy, y_numpy, 'ro')
plt.plot(X_numpy, predicted, 'b')
plt.show()
```
